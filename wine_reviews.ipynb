```python
import pandas as pd
import numpy as np
```

```python
reviews = pd.read_csv('./data/raw/wine_rev.csv')
reviews.head(5)
```

```python
reviews.info()
```

```python
reviews.describe()
```

```python
missing_data = pd.DataFrame()
```

```python
missing_variety = reviews[reviews[['variety']].isnull().any(axis=1)]
missing_country = reviews[reviews[['country']].isnull().any(axis=1)]
missing_data = pd.concat([missing_data, missing_variety, missing_country])
reviews.dropna(subset = ['variety', 'country'], inplace = True)
```

```python
missing_data.head(5)
```

```python
reviews.isna().sum() / len(reviews) * 100
```

```python
missing_price = reviews[reviews[['price']].isnull().any(axis=1)]
missing_data = pd.concat([missing_data, missing_price])
reviews.dropna(subset = ['price'], inplace = True)
```

```python
reviews.fillna({'designation': 'Unknown designation'}, inplace = True)
reviews.fillna({'region_1': 'Unknown region'}, inplace = True)
reviews.fillna({'region_2':'Unknown subregion'}, inplace = True)
reviews.fillna({'taster_name':'Unknown taster name'}, inplace = True)
reviews.fillna({'taster_twitter_handle': 'No twitter registered'}, inplace = True)
```

```python
reviews.isna().sum()
```

```python
reviews.rename(columns = {
    'Unnamed: 0': 'wine_id'
}, inplace = True)
```

```python
reviews.describe()
```

```python
reviews.info()
```

# Creating a database with clean data and a new csv

```python
import sqlite3
con = sqlite3.connect('./data/processed/clean_reviews.db')
reviews.to_sql('wine_reviews_clean', con, if_exists= 'replace', index = False)
```

```python
clean_reviews = pd.read_sql_query("SELECT * FROM wine_reviews_clean", con)
clean_reviews.info()
```
